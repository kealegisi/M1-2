```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

file_path = "DATASET.csv"  
df = pd.read_csv(file_path)

display(df.head())
```

```python
total_users_per_variation = df.groupby("variation")["user_id"].nunique()

orders_placed = df[df["final_order_status"] == "successful"].groupby("variation")["user_id"].nunique()

no_orders = total_users_per_variation - orders_placed

conversion_rate = orders_placed / total_users_per_variation

results = pd.DataFrame({
    "Variation": total_users_per_variation.index,
    "Total Users": total_users_per_variation.values,
    "Orders Placed": orders_placed.values,
    "No Orders": no_orders.values,
    "Conversion Rate": conversion_rate.values
})

display(results)
```

```python

platform_variation_users = df.groupby(["platform", "variation"])["user_id"].nunique()
platform_orders = df[df["final_order_status"] == "successful"].groupby(["platform", "variation"])["user_id"].nunique()

platform_conversion_rate = platform_orders / platform_variation_users

platform_results = pd.DataFrame({
    "Platform": platform_variation_users.index.get_level_values('platform'),
    "Variation": platform_variation_users.index.get_level_values('variation'),
    "Total Users": platform_variation_users.values,
    "Orders Placed": platform_orders.values,
    "Conversion Rate": platform_conversion_rate.values
})

display(platform_results)
```

```python

android_m1 = platform_conversion_rate.loc[("Android", 1)] if ("Android", 1) in platform_conversion_rate.index else 0
android_m2 = platform_conversion_rate.loc[("Android", 2)] if ("Android", 2) in platform_conversion_rate.index else 0
ios_m1 = platform_conversion_rate.loc[("iOS", 1)] if ("iOS", 1) in platform_conversion_rate.index else 0
ios_m2 = platform_conversion_rate.loc[("iOS", 2)] if ("iOS", 2) in platform_conversion_rate.index else 0


n_android_m1 = platform_variation_users.loc[("Android", 1)] if ("Android", 1) in platform_variation_users.index else 1
n_android_m2 = platform_variation_users.loc[("Android", 2)] if ("Android", 2) in platform_variation_users.index else 1
n_ios_m1 = platform_variation_users.loc[("iOS", 1)] if ("iOS", 1) in platform_variation_users.index else 1
n_ios_m2 = platform_variation_users.loc[("iOS", 2)] if ("iOS", 2) in platform_variation_users.index else 1


se_android = np.sqrt((android_m1 * (1 - android_m1)) / n_android_m1 + (android_m2 * (1 - android_m2)) / n_android_m2)
se_ios = np.sqrt((ios_m1 * (1 - ios_m1)) / n_ios_m1 + (ios_m2 * (1 - ios_m2)) / n_ios_m2)


z_android = (android_m1 - android_m2) / se_android if se_android != 0 else 0
z_ios = (ios_m1 - ios_m2) / se_ios if se_ios != 0 else 0


p_android = stats.norm.sf(abs(z_android)) * 2
p_ios = stats.norm.sf(abs(z_ios)) * 2


z_test_results = pd.DataFrame({
    "Platform": ["Android", "iOS"],
    "Z-Score": [float(z_android), float(z_ios)],  # Convert to float to avoid numpy issues
    "P-Value": [float(p_android), float(p_ios)]
})

display(z_test_results)
```
